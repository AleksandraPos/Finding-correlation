### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
if 'ConvertedCompYearly' in df.columns:
    df_clean = df.dropna(subset=['ConvertedCompYearly'])
    plt.figure(figsize=(12, 6))
    plt.hist(df_clean['ConvertedCompYearly'], bins=50, color='skyblue', edgecolor='black')
    plt.title('Распределение годовой компенсации (ConvertedCompYearly)', fontsize=16)
    plt.xlabel('Yearly compensation (USD)', fontsize=12)
    plt.ylabel('Частота', fontsize=12)
    plt.show()

    plt.figure(figsize=(12, 6))
    sns.kdeplot(df_clean['ConvertedCompYearly'], shade=True, color='blue', alpha=0.6)
    plt.title('График плотности распределения годовой компенсации (ConvertedCompYearly)', fontsize=16)
    plt.xlabel('Годовая компенсация (USD)', fontsize=12)
    plt.ylabel('Плотность', fontsize=12)
    plt.show()

else:
    print("Столбец 'ConvertedCompYearly' не найден в наборе данных.")
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
if 'Employment' in df.columns and 'ConvertedCompYearly' in df.columns:
    full_time_df = df[df['Employment'] == 'Employed, full-time']
    
    full_time_df_clean = full_time_df.dropna(subset=['ConvertedCompYearly'])
    
    median_compensation = full_time_df_clean['ConvertedCompYearly'].median()

    print(f"Медиана годовой компенсации для респондентов с полной занятостью: {median_compensation}")
else:
    print("Колонки 'Employment' или 'ConvertedCompYearly' не найдены в наборе данных.")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
if 'Country' in df.columns and 'ConvertedCompYearly' in df.columns:
    df_clean = df.dropna(subset=['ConvertedCompYearly', 'Country'])
    
    plt.figure(figsize=(16, 8))
    sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_clean)
    
    plt.title('Распределение компенсации по странам', fontsize=16)
    plt.xlabel('Страна', fontsize=12)
    plt.ylabel('Годовая компенсация (ConvertedCompYearly)', fontsize=12)
    plt.xticks(rotation=90) 
    plt.show()
else:
    print("Колонки 'Country' или 'ConvertedCompYearly' не найдены в наборе данных.")
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
if 'ConvertedCompYearly' in df.columns:
    df_clean = df.dropna(subset=['ConvertedCompYearly'])

    Q1 = df_clean['ConvertedCompYearly'].quantile(0.25)
    Q3 = df_clean['ConvertedCompYearly'].quantile(0.75)
    
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_no_outliers = df_clean[(df_clean['ConvertedCompYearly'] >= lower_bound) & (df_clean['ConvertedCompYearly'] <= upper_bound)]
    
    print(f"Размер исходного DataFrame: {df_clean.shape}")
    print(f"Размер DataFrame без выбросов: {df_no_outliers.shape}")
else:
    print("Колонка 'ConvertedCompYearly' не найдена в наборе данных.")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
df_corr = df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].dropna()  

corr_matrix = df_corr.corr(method='pearson')

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

plt.title('Correlation Heatmap Between Compensation, Work Experience and Job Satisfaction', fontsize=16)
plt.show()
```
